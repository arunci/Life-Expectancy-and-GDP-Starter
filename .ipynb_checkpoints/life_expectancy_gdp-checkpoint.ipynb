# Importar las Bibiotecas Necesarias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Cargar los datos
Cargamos los datos desde el archivo `all_data.csv`, usando el método `read_csv` de la biblioteca pandas. A continuación echamos un primer vistazo con el método `head()`. 

```python
df = pd.read_csv("all_data.csv")
df.head()
```

Vemos que los datos tienen cuatro columnas, con la siguiente información cada una: país, año, esperanza de vida en años y producto interior bruto (GDP).

También podemos examinar las dimensiones de los datos.

```python
df.shape
```

Los datos están constituidos por 96 filas y 4 columnas.

# Exploración de los datos

```python
# Mostrar los distintos países.
print(df.Country.unique())
```

Los seis países presentes son Chile, China, Alemania, Méjico, Estados Unidos de América y Zimbabwe.

```python
# Mostrar los años
print(df.Year.unique())
```

Los datos abarcan del año 2000 al 2015.

## Limpiar los datos

Ya en los propios nombres de las columnas hay ciertas incosistencias. Por ejemplo, el producto interior bruto viene designado con abreviatura, mientras que la columna de la esperanza de vida tiene un nombre muy largo. Por lo tanto, lo primero que vamos a hacer es designar la columna de la esperanza de vida también con abreviaturas.

```python
df = df.rename(columns={"Life expectancy at birth (years)": "LEABY"})
```

```python
df.head()
```

## Vista preliminar
### Producto Interior Bruto

```python
plt.figure(figsize=(8,6))
sns.histplot(df.GDP, kde=True)
plt.xlabel("PIB (en billones de dolares)")
```

El PIB presenta una distribución muy volcada hacia la izquierda, lo que sugiere un muy desigual reparto de la riqueza, siendo la mayoria de las naciones relativamente pobres.

### Esperanza de vida

```python
plt.figure(figsize=(8,6))
sns.histplot(df.LEABY, kde=True)
plt.xlabel("PIB (en billones de dolares)")
```

La esperanza de vida presenta un distribución muy volcada hacia la derecha, siendo casi la imagen opuesta a la del PIB.

En los anteriores apartados no hemos categorizado las observaciones por cada uno de los paises. El siguiente paso es mostrar cada una de la media de las variables por país.

```python
df_means = df.drop(["Year"], axis=1).groupby("Country").mean().reset_index()
df_means
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(y="Country", x="GDP", data=df_means)
plt.xlabel("PIB (en billones de dolares)")
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(y="Country", x="LEABY", data=df_means)
plt.xlabel("Esperanza de Vida al Nacer (años)")
```

```python
df_means_year = df.drop(["Country"], axis=1).groupby("Year").mean().reset_index()
df_means_year
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(y="LEABY", x="Year", data=df_means_year, color="navy")
plt.ylim(70, 76)
plt.xticks(rotation=45)
plt.ylabel("Esperanza de Vida al Nacer (años)")
plt.xlabel("Año")
```

```python
Vemos un leve y sostenido aumento de la esperanza de vida a lo largo de los años.
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(y="GDP", x="Year", data=df_means_year, color="navy")
plt.ylim(2e12, 6e12)
plt.xticks(rotation=45)
plt.ylabel("PIB (millones de dolares)")
plt.xlabel("Años")
```

## Diagramas de violín

Los diagramas de violín son similares a los diagramas de caja, excepto que también muestran la densidad de probabilidad de los datos en diferentes valores, generalmente suavizados por un estimador de densidad del núcleo. Un diagrama de violín incluirá todos los datos que se encuentran en un diagrama de caja: un marcador para la mediana de los datos; un cuadro o marcador que indica el rango intercuartil; y posiblemente todos los puntos de muestra, si el número de muestras no es demasiado alto.

Mientras que un diagrama de caja muestra estadísticas resumidas, como la media/mediana y rangos intercuartílicos, el diagrama de violín muestra la distribución completa de los datos. El gráfico de violín se puede utilizar en datos multimodales (más de un pico). En este caso, un gráfico de violín muestra la presencia de diferentes picos, su posición y amplitud relativa.

#### Esperanza de vida

```python
sns.violinplot(y=df.Country, x=df.LEABY)
```

Se concluye que todos los países, con la excepción de Zimbabue muestran una distribución muy estrecha; es decir muy centrada en la media, para su esperanza de vida. En cambio, la del país africano es muy amplia.

```python
df[df["Country"]== "Zimbabwe"].LEABY.max()
```

```python
df[df["Country"]== "Zimbabwe"].LEABY.min()
```

Dentro de la muestra, Zimbabue ha pasado de tener una esperanza de vida de 44.3 a 60.7, lo que supone un sorprendente aumento de más de 16 años.

#### Producto interior bruto

```python
sns.violinplot(y=df.Country, x=df.GDP)
```

Respecto al crecimiento económico, los Estados Unidos de América y China, muestran una mayor variación, mientras que el rango de Chile, Méjico y Zimbaue es muy corto. Alemania está en una posición intermedia

## PIB por año y país

```python
graphGDP = sns.FacetGrid(df, col="Country", col_wrap=3,
                      hue = "Country", sharey = False)

graphGDP = (graphGDP.map(sns.lineplot,"Year","GDP")
         .add_legend()
         .set_axis_labels("Año","PIB (billones de dolares)"))

graphGDP;
```

Se observa una tendencia al crecimiento económico en todos los paises. Destacamos la contracción económica del año 2009 que sufrieron todos los paises, con la excepción de China. También, destacamos el diferente comportamiento de las economías estadounidense y alemana. Siendo ambas muy desarrolladas, la estadounidense se recuperó mucho antes, posiblemente debido a las políticas expansivas. En cambio, en la alemana se observan sucesivos periodos de expansión y contracción.

## Esperanza de vida por país y año

```python
graphGDP = sns.FacetGrid(df, col="Country", col_wrap=3,
                      hue = "Country", sharey = False)

graphGDP = (graphGDP.map(sns.lineplot,"Year","LEABY")
         .add_legend()
         .set_axis_labels("Año","Esperanza de Vida (años)"))

graphGDP;
```

En general, la esperanza de vida a aumentado de manera continua con los años. A pesar de esa tendencia general, se observa una caida en Chile, en Méjico y en Zimbaue, que coinciden con los años de la contracción económica. 

En Alemania, a pesar de la crisis económica, no se observa una caida en la esperanza de vida. Suponemos que el ser una economía desarrollada, con un muy desarrollado estado del bienestar, protege en gran medida a sus ciudadanos, al menos en lo que se refiere a los servicios básicos de los vaivenes económicos.
