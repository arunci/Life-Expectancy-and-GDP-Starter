# Introducción
En este proyecto vamos a ver si hay alguna relación entre la riqueza de un país y la esperanza de vida de sus ciudadanos.

# Importar las Bibiotecas Necesarias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Cargar los datos
Cargamos los datos desde el archivo `all_data.csv`, usando el método `read_csv` de la biblioteca pandas. A continuación echamos un primer vistazo con el método `head()`. 

```python
df = pd.read_csv("all_data.csv")
df.head()
```

Vemos que los datos tienen cuatro columnas, con la siguiente información cada una: país, año, esperanza de vida en años y producto interior bruto (GDP).

También podemos examinar las dimensiones de los datos.

```python
df.shape
```

Los datos están constituidos por 96 filas y 4 columnas.

# Exploración de los datos

```python
# Mostrar los distintos países.
print(df.Country.unique())
```

Los seis países presentes son Chile, China, Alemania, Méjico, Estados Unidos de América y Zimbabwe.

```python
# Mostrar los años
print(df.Year.unique())
```

Los datos abarcan del año 2000 al 2015.

## Limpiar los datos

Ya en los propios nombres de las columnas hay ciertas incosistencias. Por ejemplo, el producto interior bruto viene designado con abreviatura, mientras que la columna de la esperanza de vida tiene un nombre muy largo. Por lo tanto, lo primero que vamos a hacer es designar la columna de la esperanza de vida también con abreviaturas.

```python
df = df.rename(columns={"Life expectancy at birth (years)": "LEABY"})
```

```python
df.head()
```

## Vista preliminar
### Producto Interior Bruto

```python
plt.figure(figsize=(8,6))
sns.histplot(df.GDP, kde=True)
plt.xlabel("PIB (en billones de dolares)")
```

El PIB presenta una distribución muy volcada hacia la izquierda, lo que sugiere un muy desigual reparto de la riqueza, siendo la mayoria de las naciones relativamente pobres.

### Esperanza de vida

```python
plt.figure(figsize=(8,6))
sns.histplot(df.LEABY, kde=True)
plt.xlabel("PIB (en billones de dolares)")
```

La esperanza de vida presenta un distribución muy volcada hacia la derecha, siendo casi la imagen opuesta a la del PIB.

En los anteriores apartados no hemos categorizado las observaciones por cada uno de los paises. El siguiente paso es mostrar cada una de la media de las variables por país.

```python
df_means = df.drop(["Year"], axis=1).groupby("Country").mean().reset_index()
df_means
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(y="Country", x="GDP", data=df_means)
plt.xlabel("PIB (en billones de dolares)")
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(y="Country", x="LEABY", data=df_means)
plt.xlabel("Esperanza de Vida al Nacer (años)")
```

```python
df_means_year = df.drop(["Country"], axis=1).groupby("Year").mean().reset_index()
df_means_year
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(y="LEABY", x="Year", data=df_means_year, color="navy")
plt.ylim(70, 76)
plt.xticks(rotation=45)
plt.ylabel("Esperanza de Vida al Nacer (años)")
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(y="GDP", x="Year", data=df_means_year, color="navy")
plt.ylim(2e12, 6e12)
plt.xticks(rotation=45)
plt.ylabel("PIB (millones de dolares)")
```
